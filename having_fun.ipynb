```python
import timeit
import pandas as pd
import numpy as np
import part_A
```

```python
timing_results = []
import csv
ten_thousand_list = []
thousand_list = []
hundred_list = []

# Read the CSV file
with open('data.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        # Check and append non-empty values for each column
        if len(row) > 0 and row[0].strip():  # First column
            ten_thousand_list.append(float(row[0].strip()))
        if len(row) > 1 and row[1].strip():  # Second column
            thousand_list.append(float(row[1].strip()))
        if len(row) > 2 and row[2].strip():  # Third column
            hundred_list.append(float(row[2].strip()))
```

```python
data_samples = {
    'small': hundred_list,
    'medium': thousand_list,
    'large': ten_thousand_list
}
```

```python
timing_results = [
    {'Function': func.__name__, 'Size': size, 'Time': timeit.timeit(lambda: func(data), number=10)}
    for size, data in data_samples.items()
    for func in [np.std, part_A.std_loops, part_A.std_builtin]
]
df = pd.DataFrame(timing_results)
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a bar plot
plt.figure(figsize=(6, 4))
ax = sns.barplot(x='Size', y='Time', hue='Function', data=df[:3])

# Add data labels to each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.4f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 8), 
                textcoords='offset points')

# Add labels and title
plt.xlabel('List Size')
plt.ylabel('Time (seconds)')
plt.title('Function Execution Time for Different List Sizes')
plt.show()
```
